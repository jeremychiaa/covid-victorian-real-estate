```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Read into csv file
schools_csv_file = "./Resources/dv279-schoollocations2019.csv"
schools_df = pd.read_csv(schools_csv_file)
schools_df.head()
```

```python
# Ensuring all schools are in VIC
schools_df["Postal_State"].unique()
```

```python
# Drop unwanted columns
to_drop = ["School_Status",
          "Address_Line_1",
          "Address_Line_2",
          "Address_Town",
          "Address_State",
          "Postal_Address_Line_2",
          "Full_Phone_No",
          "Address_Postcode"]

schools_df.drop(columns=to_drop, inplace=True)
```

```python
schools_df.head()
```

```python
# Store town name and postcode in a dict
suburb_postcode = dict(zip(schools_df["Postal_Town"], schools_df["Postal_Postcode"]))
suburb_postcode
```

```python
# Split LGA Name
schools_df["LGA_Name"] = schools_df.LGA_Name.apply(lambda x: pd.Series(str(x).split(" "))) 
```
