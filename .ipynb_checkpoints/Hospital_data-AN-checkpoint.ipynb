```python
#Dependencies
import pandas as pd
# Save path to data set in a variable
data_file = "Resources/Hospital_Locations.csv"
# Use Pandas to read data
df = pd.read_csv(data_file)
df.head()
```

```python
df.drop(["RoadSuffix", "CampusCode", "LocalityNa", "ServiceNam", "OpsName", "FID", "VicgovRegi"], axis=1, inplace=True)
df.head()
```

```python
# Save path to data set in a variable
data_file_2 = "Resources/suburb_postcode_df.csv"
# Use Pandas to read data
postcode_df = pd.read_csv(data_file_2)
postcode_df.head()
```

```python
hospital_df = pd.merge(df, postcode_df, on="Postcode")
hospital_df.head()
```

```python
hospital_df = hospital_df.rename(columns={"X":"Latitude", 
                                          "Y":"Longitude", 
                                          "LabelName": "Hospital Name",
                                          "Type": "Hospital Type",
                                          "StreetNum": "Street Number",
                                          "RoadName": "Road Name",
                                          "RoadType": "Road Type",
                                          "LGAName": "Local Government Area",
                                            "Postal_Town": "Suburb Name"
                                         })
hospital_df.head()
```
