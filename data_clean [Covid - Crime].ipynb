```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# loading school data to map Suburb name into Covid and Crime data
school_df = pd.read_csv("Resources\dv279-schoollocations2019.csv", encoding='ISO-8859-1')
```

```python
# rename Postal_Postcode column in school_df to postcode for merging with covid_df dataframe
school_df.rename(columns={'Postal_Postcode':'postcode'},inplace=True)
```

```python
# loading csv file
covid_df = pd.read_csv("Resources\COVID19 Data Viz Postcode data - postcode.csv")
covid_df.head(5)
```

```python
# merging to grab Postal_Town in school_df
covid_suburb_df =  pd.merge(covid_df, school_df, how = 'inner', on='postcode')
```

```python
covid_suburb_df.tail(5)
```

```python
# dropping duplicate postcode values
covid_suburb_df.drop_duplicates(subset=['postcode'])
```

```python
covid_suburb_df = covid_suburb_df[['postcode','Postal_Town','population','active','cases','rate','data_date']]
```

```python
# cleaned covid data with postcode & suburb

# rename column Postal_Town to Suburb
covid_suburb_df.rename(columns={'Postal_Town':'Suburb'})

# save cleaned data to csv file
covid_suburb_df.to_csv("Resources\covid_suburb.csv")
```
