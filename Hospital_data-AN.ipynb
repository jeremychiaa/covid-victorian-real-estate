```python
#Dependencies
import pandas as pd
# Save path to data set in a variable
data_file = "Resources/Hospital_Locations.csv"
# Use Pandas to read data
df = pd.read_csv(data_file)
df.head()
```

```python
df.drop(["RoadSuffix", "CampusCode", "LocalityNa", "ServiceNam", "OpsName", "FID", "VicgovRegi", "LGAName"], axis=1, inplace=True)
df.head()
```

```python
# Save path to data set in a variable
data_file_2 = "Resources/suburb_postcode_df.csv"
# Use Pandas to read data
postcode_df = pd.read_csv(data_file_2)
postcode_df = postcode_df.rename(columns={"Postal_Town":"Suburb"})
```

```python
# Read into created csv
suburb_lga = pd.read_csv("./output/suburb_lga_df.csv")
suburb_lga.head()
```

```python
# Merge Postcode with Local Goverment Area df using the same Suburb column
postcode_lga = pd.merge(postcode_df, suburb_lga, how="inner", on="Suburb")
postcode_lga.head()
```

```python
# Merge the Hospital df with the area info using Postcode
hospital_df = pd.merge(df, postcode_lga, on="Postcode")
hospital_df.head()
```

```python
# Rename columns of hospital df
hospital_df = hospital_df.rename(columns={"X":"Latitude", 
                                          "Y":"Longitude", 
                                          "LabelName": "Hospital Name",
                                          "Type": "Hospital Type",
                                          "StreetNum": "Street Number",
                                          "RoadName": "Road Name",
                                          "RoadType": "Road Type",
                                          "LGAName": "Local Government Area",
                                            "LGA_Name": "LGA Name"
                                         })
hospital_df.head()
```

```python
# Grouby Local Government Area
grouped_hospital = hospital_df.groupby('LGA Name')

# Find number of hospital in each LGA
hospital_count = grouped_hospital["Hospital Name"].count()

# Create a new df with hospital count
hospital_count = hospital_count.reset_index()
hospital_count

# Rename column
hospital_count = hospital_count.rename(columns={"Hospital Name": "Hospital Count per LGA"})
hospital_count.head()
```

```python
# Import pre-covid housing data
pre_covid_sales_file = "./MT_resources/vic-sales-precovid.csv"
pre_covid_sales = pd.read_csv(pre_covid_sales_file)
```

```python
# Add LGA names into pre_covid_sales
pre_covid_sales_df = pd.merge(pre_covid_sales, postcode_lga, how="inner", on="Suburb")
pre_covid_sales_df
```

```python
postcode_lga.rename(columns={"Suburb":"suburb"})
```

```python
# Merge Pre Covid data with LGA data
merged_pre_covid = pd.merge(hospital_count, pre_covid_sales, how="inner", on="Suburb")
merged_pre_covid
```

```python
# Merge hospital counts into pre-covid housing data
merged_hospital_covid = pd.merge(hospital_count, pre_covid_sales_df, how="left", on="suburb")
merged_hospital_covid
```
